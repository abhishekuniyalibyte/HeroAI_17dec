```python
# 5. Given a list of words, create a dictionary that counts how many times each word appears.
word = {
    'w1': 'apple',
    'w2': 'banana',
    'w3': 'mango',
    'w4': 'strawberry',
    'w5': 'pineapple',
    'w6': 'apple'
}
repeat = {} # empty list

for i in word.values():
    if i in repeat:
        repeat[i] +=1
    else:
        repeat[i] = 1
print(repeat)
```

```python
# 7. Sort a dictionary by its keys and values and print the sorted dictionary.

emp = {
    'name': 'abhishek',
    'profile': 'ai developer',
    'salary': 15000,
    'city': 'noida'
}

# Sort by keys (alphabetical order of keys)
sorted_emp_keys = dict(sorted(emp.items()))
print("Sorted by keys:")
print(sorted_emp_keys)

# Sort by values (converted to strings to avoid type issues)
sorted_emp_values = dict(sorted(emp.items(), key=lambda item: str(item[1])))
print("Sorted by values:")
print(sorted_emp_values)
```

```python
# 6. Build a simple menu-driven program that takes a user's choice and performs one of several actions,
# handling invalid choices with conditional statements.
a = 4
b = 5
choice = int(input('enter your choice as number: '))
if choice == 1:
    c = a + b 
    print(c)
elif choice == 2:
    c = a - b
    print(c)
elif choice == 3:
    c = a * b
    print(c)
else:
    print('invalid choice') 
```
